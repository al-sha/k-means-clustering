# Step 1: Preparing the data

```python
import pandas as pd
import numpy as np
```

```python
df  = pd.read_csv('K_Means.csv')
```

```python
df.head()
```

```python
import seaborn as sns
sns.regplot(x=df['X'], y=df['Y'])
# We now know that we can define the data into 3 clusters
```

# Step 2: Buliding the model

### Steps to build your model:
#### 1. Define: What type of model will it be? A decision tree?
#### 2. Fit: Capture patterns from provided data.
#### 3. Predict: Just what it sounds like
#### 4. Evaluate: Determine how accurate the model's predictions are

### 1. Define

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 3, init ='k-means++', max_iter=300,
                n_init = 10, random_state = 0)
```

### 2. Fitting (train)

```python
model = kmeans.fit(df)
```

### 3. Predict

```python
prediction_test = model.predict(df)
```

```python
from matplotlib import pyplot as plt
plt.scatter(df['X'] , df['Y'] , c = prediction_test )
```
